# Analyzing XGBoost SHAP Values

```python
import pandas as pd
from xgboost import XGBRegressor
import shap
shap.initjs()

%run src/columns.py
%run src/data-cleaning.py
%run src/feature-engineering.py
%run src/modeling.py
%run src/payments.py
%run src/portfolio.py
```

```python
X_train = pd.read_pickle('data/xgb_X_train.pkl.bz2', compression='bz2')
y_train = pd.read_pickle('data/xgb_y_train.pkl.bz2', compression='bz2')
```

```python
model = XGBRegressor(n_jobs=-1)
fit_model = train_model(model, X_train, y_train)
```

```python
explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X_train)
```

```python
shap.summary_plot(shap_values, X_train)
```

```python
shap.force_plot(explainer.expected_value, shap_values[-2,:], X_train.iloc[-2,:])
```

```python
shap.force_plot(explainer.expected_value, shap_values[-5,:], X_train.iloc[-5,:])
```
